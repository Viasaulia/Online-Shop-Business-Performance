```python
def item_simplifier(title):
    if 'bando' in title.lower():
        return 'Bando'
    elif 'tasbih mutiara' in title.lower():
        return 'Tasbih Mutiara'
    elif 'tasbih kayu' in title.lower():
        return 'Tasbih Kayu'
    elif 'sajadah' in title.lower():
        return 'Sajadah'
    elif 'tasbih kristal' in title.lower():
        return 'Tasbih Kristal'
    elif 'botol' in title.lower():
        return 'Botol Zamzam'
    elif 'ceko' in title.lower():
        return 'Kristal Renceng'
    elif 'renceng' in title.lower():
        return 'Kristal Renceng'
    elif 'segiempat' in title.lower():
        return 'Hijab'
    elif 'mukena' in title.lower():
        return 'Mukena'
    elif 'coklat' in title.lower():
        return 'Makanan'
    elif 'almond' in title.lower():
        return 'Makanan'
    elif 'chickpea' in title.lower():
        return 'Makanan'
    elif 'dress' in title.lower():
        return 'Dress'
    else: 
      return 'Other'
```

```python
df1['item_simp'] = df1['itemName'].apply(item_simplifier)
df1['item_simp'].value_counts()
```

```python
# check if customer ship and paid on same city, this for knowing our customer is dropshipper or not
df1['same_city'] = df1.apply(lambda x : 1 if x.shippingCity == x.billingCity else 0, axis=1)
```

### A. Metric's Breakdown (Annual and Monthly Average Revenue)

#### Customer Distribution (How Many Repeat Customer and New Customer?)

```python
## We assign each month in variabel
bulan3 = df1[(df1['purchaseTime'] >= '2022-03-01') & (df1['purchaseTime'] < '2022-04-01')]
bulan4 = df1[(df1['purchaseTime'] >= '2022-04-01') & (df1['purchaseTime'] < '2022-05-01')]
bulan5 = df1[(df1['purchaseTime'] >= '2022-05-01') & (df1['purchaseTime'] < '2022-06-01')]
bulan6 = df1[(df1['purchaseTime'] >= '2022-06-01') & (df1['purchaseTime'] < '2022-07-01')]
bulan7 =df1[(df1['purchaseTime'] >= '2022-07-01') & (df1['purchaseTime'] <= '2022-07-31')]
```

```python
cust3 = bulan3.CustId.unique()
cust4 = bulan4.CustId.unique()
cust5 = bulan5.CustId.unique()
cust6 = bulan6.CustId.unique()
cust7 = bulan7.CustId.unique()
```

```python
## We want to know how many customer repeat order in two months
repeat4= np.where(np.isin(cust3,cust4))
repeat5= np.where(np.isin(cust4,cust5))
repeat6= np.where(np.isin(cust5,cust6))
repeat7= np.where(np.isin(cust6,cust7))
total_repeat_4 = len(repeat4[0])
total_repeat_5 = len(repeat5[0])
total_repeat_6 = len(repeat6[0])
total_repeat_7 = len(repeat7[0])
```

```python
list_of_repeat = total_repeat_4, total_repeat_5,total_repeat_6,total_repeat_7
list_of_repeat = np.array(list_of_repeat)
list_of_repeat
```

```python
### this function to know how many new customer in this month

def returnNotMatches(a, b):
    notmatches = [x for x in b if x not in a]
    notmatch = len(notmatches)
    return notmatch
```

```python
new_cust4 = returnNotMatches(cust3, cust4)
new_cust5 = returnNotMatches(cust4, cust5)
new_cust6 = returnNotMatches(cust5, cust6)
new_cust7 = returnNotMatches(cust6, cust7)
new_cust = new_cust4, new_cust5, new_cust6, new_cust7
new_cust = np.array(new_cust)
new_cust
```

```python
fig, ax = plt.subplots(figsize=(17,10))
plt.title('Customer Distribution',fontsize=15)
month = ['4', '5', '6', '7']

ind = np.arange(4) 
width = 0.35  
p1 = ax.bar(ind,new_cust, width,  label='New Customer')
p2 = ax.bar(ind,list_of_repeat, width,
            bottom=new_cust, label='Repeat Customer')

plt.xlabel("Month")
plt.ylabel("Total Customer")
ax.set_xticks(ind, labels=month)
plt.legend(["New Customer", "Repeat Customer"])

ax.bar_label(p1, label_type='center')
ax.bar_label(p2, label_type='center')
ax.bar_label(p2, padding=2)

plt.show()
```
